```python
#将图片归一化并进行阈值化切片
import numpy as np
import cv2
img = cv2.imread('C:\\Users\\lpy\\Desktop\\test\\test1.png', 0)
normImg = np.zeros(img.shape, dtype=np.float32)
cv2.normalize(img,normImg,alpha=0,beta=255,norm_type=cv2.NORM_MINMAX,dtype=cv2.CV_32F)#归一化，将图片像素值缩放到0~255之间
normImg = np.array(normImg, np.uint8)
for i in range(255):#阈值化并切片，将图片中像素值低于i的点消除，高于i的保留
    ret, jyslice = cv2.threshold(normImg, i, 255, cv2.THRESH_BINARY)
    g = (i / 1) % 10
    s = (i / 10) % 10
    b = (i / 100) % 10
    cv2.imwrite('slice_' + '%d' % b + '%d' % s + '%d' % g + '.png', jyslice)
```

```python
import numpy as np#形态学操作与中值滤波，高斯滤波
import cv2
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
closing = cv2.morphologyEx(jyslice, cv2.MORPH_CLOSE, kernel)#形态学闭操作，先膨胀后腐蚀，主要用来消除图像中的小黑色区域，即将白色区域连通
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)#形态学开操作，先腐蚀后膨胀，主要用来消除图像中的小白色区域，使图像平滑
img_median = cv2.medianBlur(opening, 5)#中值滤波，又叫中值平滑，主要用来消除图像边缘的小毛刺
img_Guassian = cv2.GaussianBlur(img,(5,5),0)#高斯滤波，又叫高斯平滑，主要用来消除图片中的噪声，对高斯噪声尤为有效
```

```python
import PIL.Image as Image


# 以第一个像素为准，相同色改为透明
def transparent_back(img):
    img = img.convert('RGBA')
    L,H = img.size
    color_0 = img.getpixel((0,0))
    for h in range(H):
        for l in range(L):
            dot = (l,h)
        color_1 = img.getpixel(dot)
        if color_1 == color_0:
                color_1 = color_1[:-1] + (0,)
                img.putpixel(dot,color_1)
    return img

if __name__ == '__main__':
    img=Image.open('C:\\Users\\lpy\\Desktop\\test\\test1.png')
    img=transparent_back(img)
    img.save('C:\\Users\\lpy\\Desktop\\test\\test2.png')
```

```python
import cv2
import os
import numpy as np
from collections import Counter
from skimage import measure,color
from skimage import morphology
import matplotlib.pyplot as plt
def postprocess(img):
    # 先搞成灰度图
    img_ = np.array(img.copy(), dtype=np.int8)
    connectivity = 4
    # 连通组件
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(img_, connectivity, cv2.CV_8U)
    # labels就是由连通区域编号组成的mask 尺寸与原图对应
    labels = np.array(labels, dtype=np.float)
    # 统计各个连通区域的元素个数并排序，这里我希望得到的图像只包括三个及以下区域，于是most_common(3)
    maxnum = Counter(labels.flatten()).most_common(2)
    maxnum = sorted([x[0] for x in maxnum])
    # 制作一个包括背景和最大n个连通区域组成的Mask
    background = np.zeros_like(labels)
    if len(maxnum) == 1:
        pass
    else:
        background[labels == maxnum[1]] = 1


    # 接下来使用腐蚀-膨胀去除连通区域的毛边
  #  K_SIZE = 10
    #kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (K_SIZE, K_SIZE))
   # background = cv2.morphologyEx(background, cv2.MORPH_OPEN, kernel)
   # background = cv2.morphologyEx(background, cv2.MORPH_CLOSE, kernel)

    img[background == 0] = 0
        #  返回灰度图
    return img
srcpath="C:\\Users\\lpy\\Desktop\\test1\\test1\\"
dstpath="C:\\Users\\lpy\\Desktop\\test1\\result\\"
files = os.listdir(srcpath)
for f in files:
    imgpath=srcpath+f
    img=cv2.imread(imgpath,0)
    res=postprocess(img)
    file_name, file_extend = os.path.splitext(f)
    dst = os.path.join(os.path.abspath(dstpath), file_name + '.bmp')
    cv2.imwrite(dst, res)
```
